```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pl


gender_data = pd.read_csv("../../DataSets/pybk014master/Gender_Stat_Data.csv")

#Adolescent fertility rate (births per 1,000 women ages 15-19)
teenager_subset = gender_data.loc[gender_data['Indicator Code'] == 'SP.ADO.TFRT']

#teenager_subset

mean_teenager_1975 = teenager_subset["1975"].mean()
print ('mean',mean_teenager_1975)

median_teenager_1975 = teenager_subset["1975"].median()
print ('median',median_teenager_1975)

mode_teenager_1975 = teenager_subset["1975"].mode()
print ('mode',mode_teenager_1975)

min_teenager_1975 = teenager_subset["1975"].min()
print ('minimum',min_teenager_1975)

max_teenager_1975 = teenager_subset["1975"].max()
print('maximum', max_teenager_1975)

skew_teenager_1975 = teenager_subset["1975"].skew()
print ('skewness',skew_teenager_1975)
#nilai dari -1 sampai 1, kalau -1 puncak mepet kanan, kalau 1 puncat mepet kiri

stdev_teenager_1975 = teenager_subset["1975"].std()
print ('standard deviation',stdev_teenager_1975)
```
